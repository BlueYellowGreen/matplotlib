## 기본 사용

<br>

### 기본 그래프 그리기

<br>

`pyplot.plot()`  함수에 하나의 숫자 리스트를 입력함으로써 그래프를 그린다.

`plot()`  함수는 리스트의 값들이 y 값들이라고 가정하고, x 값을 자동으로 생성한다.

`matplotlib.pyplot.show()`  그래프를 화면에 나타나도록 한다.

<br>

```python
plt.plot([1, 2, 3, 4])
plt.show()
```

`plot()` 함수는 여러개의 인자를 받을 수 있음

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

<br>

### 스타일 지정하기

<br>

x, y 값 인자에 대해 선의 색상과 형태를 지정하는 포맷 문자열 (Format string)을 세번째 인자에 입력할 수 있다.

`ro`  red + 동그라미

`b-`  blue + 실선

참고 : https://matplotlib.org/stable/api/markers_api.html

<br>

`matplotlib.pyplot.axis()`  함수를 이용해 축의 범위 [xmin, xmax, ymin, ymax] 를 지정한다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

<br>

### 여러 개의 그래프 그리기

<br>

matplotlib 에서는 일반적으로 numpy.array 를 사용하게 된다.

하지만 numpy.array 를 사용하지 않아도 내부적으로 numpy.array 로 변환된다.

<br>

```python
t = np.arange(0, 5, 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### 레이블이 있는 데이터 사용하기

<br>

파이썬 딕셔너리와 같이 레이블이 있는 데이터를 그래프로 나타낼 수 있다.

```python
data_dict = {
    'data_x': [1, 2, 3, 4, 5],
    'data_y': [2, 3, 5, 10, 8]
}

plt.plot('data_x', 'data_y', 'r-', data=data_dict) # 어째서인지 한 번에 여러 그래프 그리는 것이 안된다.
plt.plot('data_x', 'data_x', 'g-', data=data_dict) # 그래서 이런식으로 중첩?시킨다.
plt.show()
```

## 축 레이블 설정하기

<br>

`xlabel()`, `ylabel()`  함수를 사용해서 그래프의 축에 레이블을 표시한다.

<br>

### 기본 사용

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

<br>

### 여백 지정하기

<br>

`xlabel()`, `ylabel()`  함수의  `labelpad`  파라미터는 축 레이블의 여백을 지정한다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Label', labelpad=10)
plt.ylabel('Y-Label', labelpad=20)
plt.show()
```

<br>

### 폰트 설정하기

<br>

`xlabel()`, `ylabel()` 함수의 `fontdict` 파라미터를 사용하면 축 레이블의 폰트 스타일을 설정할 수 있다.

ex) `family`, `color`, `weight`, `size`

`fontdict` 구조 참고 : https://matplotlib.org/stable/gallery/text_labels_and_annotations/text_fontdict.html

<br>

```python
font = {
    'family': 'serif',
    'color': 'b',
    'weight': 'bold',
    'size': 14
}

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Label', labelpad=10, fontdict=font)
plt.ylabel('Y-Label', labelpad=20)
plt.show()
```

<br>

### 위치 지정하기

<br>

`xlobal()` 함수의 `loc` 파라미터는 X축 레이블의 위치를 지정한다. ({'left', 'center', 'right'})

`ylabel()` 함수의 `loc` 파라미터는 y축 레이블의 위치를 지정한다. ({'bottom', 'center', 'top'})

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Label', loc='right')
plt.ylabel('Y-Label', loc='top')
plt.show()
```

<br>

## 범례 표시하기

<br>

범례(Legend)는 그래프에 데이터 종류를 표시하기 위한 텍스트이다.

`matplotlib.pyplot.legend()` 함수를 사용해서 그래프에 범례를 표시할 수 있다.

<br>

### 기본 사용

<br>

`plot()` 함수에 `label` 문자열을 지정하고 `legend()` 함수를 호출한다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Price ($)')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.legend()
plt.show()
```

<br>

### 위치 지정하기

<br>

`legend()` 함수의 `loc` 파라미터를 이용해서 범례가 표시될 위치를 설정할 수 있다.

`loc` 파라미터를 숫자 쌍 튜플로 지정하면, 해당하는 위치에 범례가 표시된다.

`loc=(0.0, 0.0)` 은 데이터 영역의 왼쪽 아래, `loc=(1.0, 1.0)` 은 데이터 영역의 오른쪽 위 위치이다.

`loc` 은 문자열로 지정할 수도 있고, 숫자 코드를 사용할 수도 있다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Price ($)')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.legend(loc=(0.0, 0.0))
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Price ($)')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.legend(loc=(0.5, 0.5))
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Price ($)')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.legend(loc=(1.0, 1.0))
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Price ($)')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.legend(loc='lower right')
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Price ($)')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.legend(loc='upper left')
plt.show()
```

<br>

### 열 개수 지정하기

<br>

`legend()` 함수의 `ncol` 파라미터는 범례에 표시될 텍스트의 열의 개수를 지정한다.

기본적으로 범례 텍스트는 1개의 열로 표시되며, `ncol=2` 로 지정하면 행으로 표시된다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Price ($)')
plt.plot([1, 2, 3, 4], [1, 2, 3, 4], label='Demand (#)')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.legend(loc='best')
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Price ($)')
plt.plot([1, 2, 3, 4], [1, 2, 3, 4], label='Demand (#)')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.legend(loc='best', ncol=2)
plt.show()
```

<br>

### 폰트 크기 지정하기

<br>

`legend()` 함수의 `fontsize` 파라미터는 범례에 표시될 폰트의 크기를 지정한다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Price ($)')
plt.plot([1, 2, 3, 4], [1, 2, 3, 4], label='Demand (#)')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.legend(loc='best', ncol=2, fontsize=14)
plt.show()
```

<br>

### 범례 테두리 꾸미기

<br>

`legend()` 함수의 `frameon` 파라미터는 범례 텍스트 상자의 테두리를 표시할지 여부를 지정한다.

`legend()` 함수의 `shadow` 파라미터를 사용해서 범례 텍스트 상자에 그림자를 표시할 수 있다.

다만 `frameon=False` 일 때 `shadow` 를 설정하는 것은 의미가 없다.

<br>

이외에도 `facecolor`, `edgecolor`, `borderpad`, `labelspacing` 과 같은 다양한 파라미터가 있다.

참고 : https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Price ($)')
plt.plot([1, 2, 3, 4], [1, 2, 3, 4], label='Demand (#)')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.legend(loc='best', ncol=2, fontsize=10, frameon=True, shadow=True)
plt.show()
```

<br>

## 축 범위 지정하기

<br>

`matplotlib.pyplot` 모듈의 `xlim()`, `ylim()`, `axis()` 함수를 사용하면 X, Y 축이 표시되는 범위를 지정할 수 있다.

`xlim()` : X축이 표시되는 범위를 지정하거나 반환한다. `(xmin, xmax)` or `((xmin, xmax))` or `([xmin, max])`

`ylim()` : Y축이 표시되는 범위를 지정하거나 반환한다. `(ymin, ymax)` or `((ymin, ymax))` or `([ymin, ymax])`

`axis()` : X, Y축이 표시되는 범위를 지정하거나 반환한다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim(0, 5)
plt.ylim(0, 20)
plt.show()
```

<br>

### 기본사용 - axis()

<br>

`axis()` 함수에 `[xmin, xmax, ymin, ymax]` 의 형태로 X, Y축의 범위를 지정할 수 있다.

입력값이 없으면 데이터에 맞게 자동으로 범위를 지정한다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0, 5, 0, 20])
plt.show()
```

<br>

### 옵션 지정하기

<br>

`axis()` 함수에는 축에 관한 다양한 옵션을 제공한다.

`on`, `off`, `equal`, `scaled`, `tight`, `auto`, `normal`, `image`, `square`

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='square')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis('square')
plt.legend()
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='scaled')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis('scaled')
plt.legend()
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='off')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis('off')
plt.legend()
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='equal')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis('equal')
plt.legend()
plt.show()
```

<br>

### 축 범위 얻기

<br>

`xlim()`, `ylim()`, `axis()` 모두 범위를 반환한다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='equal')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

axis_range = plt.axis('equal')
print(axis_range)

plt.legend()
plt.show()
```

### 기본 사용

<br>

Solid : `-`

Dashed : `--`

Dotted : `:`

Dash-dot : `-.`

<br>

```python
plt.plot([1, 2, 3], [4, 4, 4], '-', label='Solid')
plt.plot([1, 2, 3], [3, 3, 3], '--', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], ':', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-.',label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
```

## 그래프 영역 채우기

<br>

`matplotlib.pyplot` 모듈에서 그래프의 영역을 채우는 아래의 세가지 함수를 소개한다.

`fill_between()` : 두 수평 방향의 곡선 사이를 채운다.

`fill_betweenx()` : 두 수직 방향의 곡선 사이를 채운다.

`fill()` : 다각형 영역을 채운다.

<br>

### 기본 사용 - fill_between()

<br>

`fill_between()` 함수에 `x[start:end]`, `y[start:end]` 를 순서대로 입력하면 해당 영역이 채워진다.

<br>

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], alpha=0.5)
plt.show()
```

<br>

### 기본 사용 - fill_betweenx()

<br>

`fill_betweenx()` 함수에 `y[start:end]`, `x[start:end]` 를 순서대로 입력하면 해당 영역이 채워진다.

<br>

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y,)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_betweenx(y[1:3], x[1:3], alpha=0.5)
plt.show()
```

<br>

### 두 그래프 사이 영역 채우기

<br>

두 개의 그래프 사이 영역을 채우기 위해서 두 개의 y값의 리스트를 입력한다.

<br>

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [3, 7, 12, 19]

plt.plot(x, y1, x, y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='g', alpha=0.3)
plt.show()
```

<br>

### 다각형 영역 채우기 - fill()

<br>

`fill()` 함수에 x, y 값의 리스트를 입력하면, 다각형 영역을 자유롭게 지정해서 채울 수 있다.

<br>

```python
plt.axis([0, 5, 0, 5])
plt.fill([1, 2, 3], [3, 3, 2], color='r', alpha=0.2)
plt.show()
```

## 축 스케일 지정하기

<br>

`matplotlib.pyplot` 모듈의 `xscale()`, `yscale()` 함수를 사용해서 그래프의 축 스케일을 지정할 수 있다.

축은 기본적으로 `linear` 스케일로 표시되지만 `log`, `symlog`, `logit` 으로 변경할 수 있다.

<br>

### X축 스케일 지정하기

<br>

```python
x = np.linspace(-10, 10, 100)
y = x ** 3

plt.plot(x, y)
plt.show()

plt.plot(x, y, label='log')
plt.xscale('log')
plt.legend(loc='best')
plt.show()

plt.plot(x, y, label='symlog')
plt.xscale('symlog')           # symlog : Symmetrical Log Scale
plt.legend(loc='best')
plt.show()

plt.plot(x, y, label='logit')
plt.xscale('logit')
plt.legend(loc='best')
plt.show()
```

<br>

### Y축 스케일 지정하기

<br>

```python
x = np.linspace(-10, 10, 100)
y = x ** 3

plt.plot(x, y)
plt.show()

plt.plot(x, y, label='log')
plt.yscale('log')
plt.legend(loc='best')
plt.show()

plt.plot(x, y, label='symlog')
plt.yscale('symlog')           # symlog : Symmetrical Log Scale
plt.legend(loc='best')
plt.show()

plt.plot(x, y, label='logit')
plt.yscale('logit')
plt.legend(loc='best')
plt.show()
```

## 그리드 설정하기

<br>

`matplotlib.pyplot` 모듈의 `grid()` 함수를 이용해서 그래프에 다양하게 그리드를 설정할 수 있다.

<br>

### 기본 사용

<br>

`plt.grid(True)` 같이 설정하면, x, y 축에 대해 그리드가 표시된다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.grid(True)
plt.show()
```

<br>

### 축 지정하기

<br>

`axis` 파라미터로 원하는 축의 방향으로만 그리드를 표시할 수 있다. `{'both', 'x', 'y'}`

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True, axis='x')
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True, axis='y')
plt.show()
```

<br>

### 스타일 설정하기

<br>

`color`, `marker`, `alpha`, `linestyle` 파라미터를 사용해서 그리드 선의 스타일을 설정할 수 있다.

또한 `which` 파라미터를 `major`, `minor`, `both` 등으로 사용하면 주눈금, 보조눈금에 각각 그리드를 표시할 수 있다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True, axis='y', color='red', linestyle='--', linewidth=2, alpha=0.1)
plt.show()
```

## 눈금 표시하기

<br>

틱(Tick)은 그래프의 축에 간격을 구분하기 위해 표시하는 눈금이다.

`matplotlib.pyplot` 모듈의 `xticks()`, `yticks()`, `tick_params()` 함수를 이용해서 그래프에 눈금을 표시할 수 있다.

<br>

### 기본 사용

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xticks([1, 2, 3, 4])
plt.yticks(range(1, 20))
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xticks([1, 2, 3, 4])
plt.yticks([1, 5, 10, 15, 20])
plt.show()
```

<br>

### 눈금 레이블 지정하기

<br>

`labels` 파라미터를 사용하면 눈금 레이블을 문자열의 형태로 지정할 수 있다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xticks([1, 2, 3, 4], labels=['label1', 'label2', 'label3', 'label4'])
plt.yticks([1, 5, 10, 15, 20], labels=['A', 'B', 'C', 'D', 'E'])
plt.show()
```

<br>

### 눈금 스타일 설정하기

<br>

`tick_params()` : 함수를 사용하면 눈금의 스타일을 다양하게 설정할 수 있다.

`axis` : 설정이 적용될 축을 지정한다. `{'x', 'y', 'both'}`

`direction` : `in`, `out` 으로 설정하면 눈금이 안/밖으로 표시된다. `{'in', 'out', 'inout'}`

`length` : 눈금의 길이를 지정한다.

`pad` : 눈금과 레이블과의 거리를 지정한다.

`labelsize` : 레이블의 크기를 지정한다.

`labelcolor` : 레이블의 색상을 지정한다.

`top/bottom/left/right` : 이것을 `True/False` 로 지정하면 눈금이 표시될 위치를 선택할 수 있다.

`width` : 눈금의 너비를 지정한다.

`color` : 눈금의 색상을 지정한다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xticks([1, 2, 3, 4], labels=['label1', 'label2', 'label3', 'label4'])
plt.yticks([1, 5, 10, 15, 20], labels=['A', 'B', 'C', 'D', 'E'])
plt.tick_params(axis='x', direction='in', length=3, pad=6, labelsize=14, labelcolor='green', top=True)
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xticks([1, 2, 3, 4], labels=['label1', 'label2', 'label3', 'label4'])
plt.yticks([1, 5, 10, 15, 20], labels=['A', 'B', 'C', 'D', 'E'])
plt.tick_params(axis='y', direction='inout', length=10, pad=6, labelsize=14, labelcolor='red', width=2, color='r')
plt.show()
```

<br>

## 타이틀 설정하기

<br>

`matplotlib.pyplot` 모듈의 `title()` 함수를 이용해서 그래프의 타이틀을 설정할 수 있다.

<br>

### 기본 사용

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('Simple Graph')
plt.show()
```

<br>

### 위치와 오프셋 지정하기

<br>

`plt.title()` 함수의 `loc` 파라미터를 `right` 로 설정하면, 타이틀이 그래프의 오른쪽 위에 나타나게 된다.

&#10140;  `{'left', 'right', 'center'}` 중 선택 가능. (디폴트는 `center`)

`pad` 파라미터는 타이틀과 그래프와의 간격을 포인트 단위로 설정한다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('Simple Graph', loc='center', pad=5)
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('Simple Graph', loc='right', pad=10)
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('Simple Graph', loc='left', pad=20)
plt.show()
```

<br>

### 폰트 지정하기

<br>

`fontdict` 파라미터에 딕셔너리 형태로 폰트 스타일을 설정할 수 있다.

`fontsize` : 포인트 단위의 숫자를 입력하거나 `smaller`, `x-large` 등의 상대적인 설정을 할 수 있다.

`fontweight` : `normal`, `bold`, `heavy`, `light`, `ultrabold`, `ultralight` 같이 설정할 수 있다.

<br>

```python
font = {
    'fontsize': 16,
    'fontweight': 'bold'
}

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('Simple Graph', fontdict=font)
plt.show()
```

## 수평선/수직선 표시하기

<br>

그래프의 특정 위치에 수직선/수평선을 표시하기 위해서 `matplotlib.pyplot` 모듈은 아래의 네가지 함수를 지원한다.

`axhline()` : 축을 따라 수평선을 표시한다.

`axvline()` : 축을 따라 수직선을 표시한다.

`hlines()` : 지정한 점을 따라 수평선을 표시한다.

`vlines()` : 지정한 점을 따라 수직선을 표시한다.

<br>

### 수평선 그리기 - axhline(), hlines()

<br>

`axhline()` 함수의 첫 번째 인자는 y 값으로서 수평선의 위치가 된다.

두, 세 번째 인자는 `xmin`, `xmax` 값으로서 0에서 1사이의 값을 입력한다. (0 : left, 1 : right)

`hlines()` 함수에 `y`, `xmin`, `xmax` 를 순서대로 입력하명 `(xmin, y)` 에서 `(xmax, y)` 를 따라 수평선을 표시한다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axhline(8, 0.1, 0.9, color='red', linestyle='--', linewidth=2)
plt.hlines(7, 1, 4, color='green', linestyle='-', linewidth=1)
plt.show()
```

<br>

### 수직선 그리기 - axvline(), vlines()

<br>

`axvline()` 함수의 첫 번째 인자는 x 값으로서 수직선의 위치가 된다.

두, 세 번째 인자는 `ymin`, `ymax` 값으로서 0에서 1사이의 값을 입력한다. (0 : bottom, 1 : top)

`vlines()` : 함수에 x, `ymin`, `ymax` 를 순서대로 입력하면, `(x, ymin)` 에서 `(x, ymax)` 를 따라 수평선을 표시한다.

<br>

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axvline(2.5, 0.1, 0.9, color='red', linestyle='--', linewidth=1)
plt.vlines(2, 1, 16, color='green', linestyle='-', linewidth=1)
plt.show()
```

## 막대 그래프 그리기

<br>

막대 그래프는 범주가 있는 데이터 값을 직사각형의 막대로 표현하는 그래프이다.

`matplotlib.pyplot` 모듈의 `bar()` 함수를 이용해서 막대 그래프를 표현할 수 있다.

<br>

### 기본 사용

<br>

`plt.bar()` 함수에 x, y 값을 입력해주고, `xticks()` 에 x 와 눈금을 입력해주면 된다.

<br>

```python
x = [1, 2, 3]
x_label = ['val1', 'val2', 'val3']
y = [300, 400, 500]

plt.bar(x, y)
plt.xticks(x, x_label)
plt.show()
```

<br>

### 색상 시정하기

<br>

`plt.bar()` 함수의 `color` 파라미터를 사용해서 막대의 색상을 지정할 수 있다.

<br>

```python
x = [1, 2, 3]
x_label = ['val1', 'val2', 'val3']
y = [300, 400, 500]

plt.bar(x, y, color=['r', 'g', 'b'])
plt.xticks(x, x_label)
plt.show()
```

<br>

### 막대 폭 지정하기

<br>

`plt.bar()` 함수의 `width` 파라미터를 이용하여 막대의 폭을 지정한다. (디폴트는 0.8 이다.)

<br>

```python
x = [1, 2, 3]
x_label = ['val1', 'val2', 'val3']
y = [300, 400, 500]

plt.bar(x, y, width=0.4)
plt.xticks(x, x_label)
plt.show()
```

<br>

### 스타일 꾸미기

<br>

`align` 은 눈금과 막대의 위치를 조절한다. (`center`, `edge` 디폴트=`center`)

`edgecolor` 는 막대 테두리 색, `linewidth` 는 테두리의 두께를 지정한다.

`tick_label` 을 리스트 또는 array 형태로 지정하면, tick 에 문자열을 순서대로 나타낼 수 있다.

<br>

```python
x = [1, 2, 3]
x_label = ['val1', 'val2', 'val3']
y = [300, 400, 500]

plt.bar(x, y, width=0.4, align='edge', edgecolor='r', linewidth='3', tick_label=x_label)
plt.show()
```

## 산점도 그리기

<br>

산점도는 두 변수의 상관 관계를 직교 좌표계의 평면에 점으로 표현하는 그래프이다.

`matplotlib.pyplot` 모듈의 `scatter()` 함수를 이용하면 산점도를 그릴 수 있다.

<br>

### 기본 사용

<br>

`scatter()` 함수에 x, y 값을 입력하면 된다.

<br>

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)

plt.scatter(x, y)
plt.show()
```

<br>

### 색상과 크기 지정하기

<br>

`scatter()` 함수의 `s`, `c` 파라미터는 각각 마커의 크기와 색상을 지정한다.

마커의 크기는 size ** 2 의 형태로 지정한다.

마커의 색상은 데이터의 길이와 같은 크기의 숫자 시퀀스 또는 rgb, 그리고 Hex code 색상을 입력해서 지정한다.

<br>

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)

area = (30 * np.random.rand(n)) ** 2 # plot(markersize = 30 * np.random.rand(n)) 와 동일하다.
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors)
plt.show()
```

<br>

### 투명도와 컬러맵 설정하기

<br>

`alpha` 파라미터는 마커의 투명도를 지정한다. (0 ~ 1)

`cmap` 파라미터에 컬러맵에 해당하는 문자열을 지정할 수 있다.

<br>

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)

area = (30 * np.random.rand(n)) ** 2
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors, alpha=0.4, cmap='Spectral')
plt.colorbar() # 이것을 통해 사용하는 색상의 범위를 확인할 수 있다.
plt.show()
```

<br>

## 히스토그램 그리기

<br>

히스토그램은 도수분포표를 그래프로 나타낸 것이다.

`matplotlib.pyplot` 모듈의 `hist()` 함수를 이용해서 히스토그램을 그린다.

<br>

### 기본 사용

<br>

`hist()` 함수에 리스트의 형태로 값들을 입력하면 된다.

<br>

```python
np.random.seed(0)

val = np.array(np.random.rand(50) * 30 + 5, dtype=np.int8)

plt.hist(val)
plt.show()
```

<br>

### 구간 개수 지정하기

<br>

`hist()` 함수의 `bins` 파라미터는 히스토그램의 가로축 구간의 개수를 지정한다.

구간의 개수에 따라 히스토그램 분포의 평태가 달라질 수 있기 때문에 적절한 개수를 지정해야 한다.

<br>

```python
np.random.seed(0)

val = np.array(np.random.rand(50) * 30 + 5, dtype=np.int8)

plt.hist(val, label='bins=10')
plt.hist(val, bins=30, label='bins=30')
plt.hist(val, bins=50, label='bins=50')
plt.legend()
plt.show()
```

<br>

### 누적 히스토그램 그리기

<br>

`cumulative` 파라미터를 `True` 로 지정하면 누적 히스토그램을 나타낸다. (디폴트=False)

<br>

```python
np.random.seed(0)

val = np.array(np.random.rand(50) * 30 + 5, dtype=np.int8)

plt.hist(val, bins=30, cumulative=True, label='True')
plt.hist(val, bins=30, label='False')
plt.legend()
plt.show()
```

<br>

### 히스토그램 종류 지정하기

<br>

`histtype` 은 히스토그램의 종류를 지정한다.

`{'bar', 'barstacked', 'step', 'stepfilled'}` 중에서 선택할 수 있으며, 디폴트는 'bar' 이다.

<br>

```python
np.random.seed(0)

val = np.array(np.random.rand(50) * 30 + 5, dtype=np.int8)
val2 = np.array(np.random.rand(50) * 50 + 1, dtype=np.int8)

plt.hist((val, val2), histtype='bar')
plt.title('bar')
plt.figure()

plt.hist((val, val2), histtype='barstacked')
plt.title('barstacked')
plt.figure()

plt.hist((val, val2), histtype='step')
plt.title('step')
plt.figure()

plt.hist((val, val2), histtype='stepfilled')
plt.title('stepfilled')
plt.show()
```

<br>

### numpy 난수의 분포 나타내기

<br>

`density=True` 로 설정하면, 밀도함수가 되어서 막대의 아래 면적이 1이 된다.

<br>

```python
np.random.seed(0)

a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')

plt.show()
```

## 에러바 표시하기

<br>

에러바는 데이터의 편차를 표시하기 위한 그래프 형태이다.

`matplotlib.pyplot` 모듈의 `errorbar()` 함수를 이용해서 그래프에 에러바를 나타낼 수 있다.

<br>

### 기본 사용

<br>

`errorbar()` 함수에 x, y 값들과 함께 데이터의 편차를 나타내는 리스트인 yerr 를 입력한다.

리스트의 값들은 데이터 포인트의 위/아래 대칭인 오차로 표시된다.

<br>

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
yerr = [1, 5, 1, 5]

plt.errorbar(x, y, yerr=yerr)
plt.show()
```

<br>

### 비대칭 편차 나타내기

<br>

데이터 포인트를 기준으로 비대칭인 편차를 표시하려면, (2, N) 형태의 값들을 입력하면 된다. (N: 데이터 개수)

첫 번째 튜플의 값들은 아래 방향 편차, 두 번째 튜플의 값들은 위 방향 편차를 나타낸다.

<br>

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
yerr = [(1, 3, 1, 3), (3, 1, 3, 1)]

plt.errorbar(x, y, yerr=yerr)
plt.show()
```

<br>

### 상한/하한 기호 표시하기

<br>

`uplims`, `lolims` 를 사용해서 상한/하한 기호를 선택적으로 표시할 수 있다.

예를 들어, `uplims` 가 True 이고 `lolims` 가 False 라면, 이 값이 상한값임을 의미해서 아래 방향의 화살표가 표시된다.

<br>

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
yerr = [(1, 3, 1, 3), (3, 1, 3, 1)]

upperlimits = [True, True, False, False]
lowerlimits = [False, False, False, True]

plt.errorbar(x, y, uplims=upperlimits, lolims=lowerlimits, yerr=yerr)
plt.show()
```

## 파이 차트 그리기

<br>

파이 차트는 범주별 구성 비율을 원형으로 표현한 그래프이다.

`matplotlib.pyplot` 모듈의 `pie()` 함수를 이용해서 파이 차트를 그릴 수 있다.

<br>

### 기본 사용

<br>

`ratio` 와 `labels` 로 영역의 비율과 이름을 정한다.

`autopct` 는 차트 안에 표시될 숫자의 형식을 지정한다.

<br>

## 여러 개의 그래프 그리기

<br>

`matplotlib.pyplot` 모듈의 `subplot()` 함수는 여러 개의 그래프를 하나의 그림에 나타내도록 한다.

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)                # nrows=2, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)                # nrows=2, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

<br>

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.ylabel('Damped oscillation')
plt.xticks(visible=False)

ax2 = plt.subplot(2, 1, 2, sharex=ax1)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

## 객체 지향 인터페이스

matplotlib 은 그래프를 다루는 두 가지의 인터페이스를 제공하는데, 첫 번째는 MATLAB 스타일로 `pyplot` 모듈을 사용하는 방식이고, 두 번째는 객체 지향 인터페이스이다.

<br>

### plt.subplots() 사용하기

<br>

`matplotlib.pyplot` 모듈은 `subplots()` 라는 유용한 함수를 제공한다.

`subplots()` 함수를 호출하면 figure (fig) 와 subplot (ax) 객체를 생성해서 튜플의 형태로 반환한다.

<br>

```python
fig, ax = plt.subplots()
plt.show()
```

<br>

### 행과 열 설정하기 (nrows, ncols)

<br>

`plt.subplots(nrows, ncols)` 의 형태로 행과 열의 개수를 지정할 수 있다.

만약 지정하지 않으면 기본적으로 행과 열의 개수는 모두 1 이다.

<br>

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 8))
plt.show()
```

<br>

### X, Y 축 공유하기 (sharex, sharey)

<br>

`sharex=True`, `sharey=True` 로 설정함으로써 중복된 축을 한 번만 표시하도록 할 수 있다.

`sharex`, `sharey` 에 True, False 외에도 `'all'`, `'none'`, `'row'`, `'col'` 등을 지정할 수 있다.

<br>

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 8), sharex=True, sharey=True)
plt.show()
```

<br>

### 그래프 그리기

<br>

`plt.subplots()` 이 반환하는 ax 는 matplotlib 의 axes 클래스의 인스턴스이다.

행과 열을 2, 2 로 지정하면 ax 는 2 x 2 형태를 갖는 numpy array 이다.

<br>

```python
x = np.arange(1, 5)

fig, ax = plt.subplots(2, 2, figsize=(8, 8), sharex=True, sharey=True, squeeze=True)
ax[0][0].plot(x, np.sqrt(x))     # left-top
ax[0][1].plot(x, x)              # right-top
ax[1][0].plot(x, -x+5)           # left-bottom
ax[1][1].plot(x, np.sqrt(-x+5))  # right-bottom

plt.show()
```

<br>

### 제목과 범례 표시하기

<br>

`set_title()` 과 `legend()` 를 이용해서 각각의 그래프에 제목과 범례를 설정할 수 있다.

<br>

```python
x = np.arange(1, 5)

fig, ax = plt.subplots(2, 2, figsize=(8, 8), sharex=True, sharey=True, squeeze=True)
ax[0][0].plot(x, np.sqrt(x), 'gray', linewidth=3, label='y=np.sqrt(x)')
ax[0][0].set_title('Graph 1')
ax[0][0].legend()
ax[0][1].plot(x, x, 'g^-', markersize=10, label='y=x')
ax[0][1].set_title('Graph 2')
ax[0][1].legend(loc='upper left')
ax[1][0].plot(x, -x+5, 'ro--', label='y=-x+5')
ax[1][0].set_title('Graph 3')
ax[1][0].legend(loc='lower left')
ax[1][1].plot(x, np.sqrt(-x+5), 'b.-.', label='y=np.sqrt(-x+5)')
ax[1][1].set_title('Graph 4')
ax[1][1].legend(loc='upper center')

plt.show()
```

## 축 위치 조절하기

<br>

matplotlib 의 기분적인 레이아웃은 그래프 상하좌우의 네 방향에 데이터 영역을 나타내는 직선이 그려지는 형태이다.

<br>

### 기본 사용

<br>

`ax.spine[]` 는 해당 축의 spine 을 가리킨다. (spine : 데이터 영역의 경계를 나타내는 선)

`set_position()` 는 데이터 영역의 어느 부분에 위치시킬지 의미한다.

`set_position(('data', 0))` 과 같이 특정 데이터 좌표의 위치에 직선을 표시할 수 있다.

`set_visible()` 은 spine 이 표시될지 말지를 결정한다.

<br>

```python
fig, ax = plt.subplots()

ax.set_title('Mean Squared Error', pad=20)
ax.set_xlim(-3, 3)
ax.set_ylim(0, 3)
ax.set_xticks([-3, -2, -1, 0, 1, 2, 3])
ax.set_yticks([1, 2, 3])

ax.spines['left'].set_position('center')       # 왼쪽 축을 가운데 위치로 이동
ax.spines['right'].set_visible(False)          # 오른쪽 축을 보이지 않도록
ax.spines['top'].set_visible(False)            # 위 축을 보이지 않도록
ax.spines['bottom'].set_position(('data', 0))  # 아래 축을 데이터 0의 위치로 이동
ax.tick_params('both', length=0)               # Tick의 눈금 길이 0

x = np.linspace(-3, 3, 100)
ax.set_xlabel('y_true - y_pred', fontdict={'fontsize': 14}, labelpad=10)
ax.plot(x, x**2, color='#4799FF', linewidth=5)

plt.show()
```

## 이중 Y축 표시하기

<br>

두 종류의 데이터를 동시에 하나의 그래프에 표시하기 위해 이중 축을 표시할 수 있다.

<br>

### 기본 사용

<br>

`plt.subplots()` 로 반환된 ax 인스턴스를 이용하여 `ax.plot(x, y)` 는 첫 번째 축에 x, y 데이터를 나타낸다.

`ax.twinx()` 는 ax 와 x 축을 공유하는 새로운 axes 객체를 만든다.

`ax2.plot(x, y2)` 는 새로운 axes 객체에 x, y2 데이터를 나타낸다.

<br>

```python
x = np.arange(0, 3)
y1 = x + 1
y2 = -x - 1

fig, ax = plt.subplots()
ax.plot(x, y1, color='green')

ax2 = ax.twinx()
ax2.plot(x, y2, color='red')

plt.show()
```

<br>

### 축 레이블 표시하기

<br>

`set_xlabel()`, `set_ylabel()` 메서드는 각 축에 대한 레이블을 표시하도록 한다.

<br>

```python
x = np.arange(0, 3)
y1 = x + 1
y2 = -x - 1

fig, ax = plt.subplots()
ax.plot(x, y1, color='green')
ax.set_ylabel('y1 Axis')

ax2 = ax.twinx()
ax2.plot(x, y2, color='red')
ax2.set_ylabel('y2 Axis')

plt.show()
```

<br>

### 범례 표시하기

<br>

각 축의 데이터 곡선에 대한 범례를 나타내기 위해 `legend()` 메서드를 사용한다.

<br>

```python
x = np.arange(0, 3)
y1 = x + 1
y2 = -x - 1

fig, ax = plt.subplots()
ax.plot(x, y1, color='green', label='y1')
ax.legend()

ax2 = ax.twinx()
ax2.plot(x, y2, color='red', label='y2')
ax2.legend()

plt.show()
```

```python
x = np.arange(0, 3)
y1 = x + 1
y2 = -x - 1

fig, ax = plt.subplots()
legend1 = ax.plot(x, y1, color='green', label='y1')

ax2 = ax.twinx()
legend2 = ax2.plot(x, y2, color='red', label='y2')

legends = legend1 + legend2
labels = [l.get_label() for l in legends]
ax.legend(legends, labels, loc='upper center')

plt.show()
```

## 두 종류의 그래프 그리기

<br>

앞서 소개한 이중 Y축 표시하기를 이용해서 두 종류의 그래프를 하나의 그래프 영역에 표현한다.

<br>

### 기본 사용

<br>

우선 `ax.twinx()` 로 X축을 공유하는 이중 Y축을 만들고, `ax.plot()` 과 `ax2.bar()` 를 사용해서 y1, y2 데이터를 그릴 수 있다.

<br>

```python
# 1. 데이터 준비
x = np.arange(2020, 2027)
y1 = np.array([1, 3, 7, 5, 9, 7, 14])
y2 = np.array([1, 3, 5, 7, 9, 11, 13])

# 2. 그래프 그리기
fig, ax = plt.subplots()

ax.plot(x, y1, '-', color='green', linewidth=2, alpha=0.7)
ax.set_ylim(0, 16)
ax.set_ylabel('Price ($)')

ax2 = ax.twinx()
ax2.bar(x, y2, color='red', alpha=0.7, width=0.5)
ax2.set_ylim(0, 16)
ax2.set_ylabel(r'Demand ($\times10^6$)')

plt.show()
```

## 박스 플롯 그리기

<br>

박스 플롯 또는 박스-위스커 플롯은 수치 데이터를 표현하는 하나의 방식이다.

일반적으로 전체 데이터로부터 얻어진 아래의 다섯 가지 요약 수치를 사용해서 그려진다.

- 최솟값, Q1, Q2, Q3, 최댓값

<br>

### 기본 사용

<br>

`ax.boxplot()` 은 데이터 array 로부터 얻어진 요약 수치를 박스 형태로 나타낸다.

가운데 박스 모양으로부터 그려지는 중심선을 수염(Whisker) 라고 한다.

`boxplot()` 함수는 기본적으로 위쪽 경계로부터 Q3 + whis*(Q3-Q1) 까지, 아래쪽 경계로부터 Q1 - whis*(Q3-Q1) 까지 나타낸다.

whis 의 디폴트 값은 1.5이며, whis 외부 데이터 포인트(Outlier) 는 개별적으로 표시된다.

<br>

```python
# 1. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 2. 그래프 그리기
fig, ax = plt.subplots(figsize=(5, 5))

ax.boxplot([data_a, data_b, data_c])
ax.set_ylim(-10.0, 10.0)
plt.show()
```

<br>

## 노치 표시하기

<br>

`ax.boxplot()` 의 `notch` 파라미터를 True 로 설정하면, 중앙값의 95% 신뢰 구간을 노치 형태로 표시한 Notched Boxplot 을 나타낸다.

<br>

```python
# 1. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 2. 그래프 그리기
fig, ax = plt.subplots(figsize=(5, 5))

ax.boxplot([data_a, data_b, data_c], notch=True, whis=2.5) # whis=2.5 에 의해 whis 가 길어지고, 이에 outlier 가 표시되지 않는다.
ax.set_ylim(-10.0, 10.0)
plt.show()
```

## 바이올린 플롯 그리기

<br>

바이올린 플롯은 데이터의 분포와 범위를 한 눈에 보기 쉽게 나타내는 그래프 형식이다.

박스 플롯과 비슷하지만, 더 실제에 가까운 분포를 알 수 있다는 장점이 있다.

<br>

### 기본 사용

<br>

`ax.violinplot()` 은 주어진 데이터 array 분포를 바이올린의 형태로 시각화한다.

`potitions` 파라미터는 바이올린 플롯의 x 위치를 지정한다. (지정하지 않으면 1, 2, 3... 의 순서로 그래프가 그려진다.)

`ax.set_xticks([...])` 를 통해 x 축의 눈금을 지정했고, `[...]` 위치에 그래프가 그려진다.

`showmeans` 파라미터를 True 로 설정하면 평균값을 표시한다. (디폴트=False)

`showextrema` 파라미터를 True 로 설정하면 최대/최소값을 표시한다. (디폴트=True)

`showmedians` 파라미터를 True 로 설정하면 중앙값을 표시한다. (디폴트=False)

`quantiles` 파라미터는 데이터 분포의 분위수를 표시한다. (0.0 ~ 1.0)

<br>

```python
# 1. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 2. 그래프 그리기
fig, ax = plt.subplots(figsize=(5, 5))

ax.violinplot([data_a, data_b, data_c], positions=[2, 3, 4], showmeans=True, quantiles=[[.25], [.75], []])
ax.set_ylim(-10.0, 10.0)
ax.set_xticks([1, 2, 3, 4, 5])

plt.show()
```

<br>

### 색상 지정하기

<br>

`ax.violinplot()` 은 바이올린 플롯의 각 구성 요소에 해당하는 collection 인스턴스의 리스트를 딕셔너리의 형태로 반환한다.

반환한 딕셔너리는 키를 포함하고 있다.

- `bodies` : 각각의 바이올린 분포의 채워진 영역을 포함하는 PolyCollection 인스턴스의 리스트

- `cmeans` : 바이올린 분포의 평균값을 표시하는 LineCollection 인스턴스

- `cmins` : 바이올린 분포의 최솟값을 표시하는 LineCollection 인스턴스

- `cmaxes` : 바이올린 분포의 최댓값을 표시하는 LineCollection 인스턴스

- `cbars` : 바이올린 분포의 중심을 표시하는 LineCollection 인스턴스

- `cmedians` : 바이올린 분포의 중앙값을 표시하는 LineCollection 인스턴스

- `cquantiles` : 바이올린 분포의 분위값을 표시하는 LineCollection 인스턴스

<br>

```python
# 1. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 2. 그래프 그리기
fig, ax = plt.subplots(figsize=(5, 5))

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True)
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])

violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')

violin['cbars'].set_edgecolor('gray')
violin['cmaxes'].set_edgecolor('gray')
violin['cmins'].set_edgecolor('gray')
violin['cmeans'].set_edgecolor('gray')

plt.show()
```
